```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,4,3]

plt.plot(x, y, label = "dados")
plt.ylabel("eixo y")
plt.xlabel("eixo x")
plt.title("Gr√°fico")
plt.yticks([-1,3,5,9,11])
plt.xticks([0,2,4,6,8,9])
plt.show()
```

```python
plt.plot (x,y,linestyle="solid")
plt.show()
```

```python
plt.plot (x,y,linestyle="dashed")
plt.show()
```

```python
plt.plot (x,y,linestyle="--")
plt.show()
```

```python
plt.plot (x,y,linestyle=":")
plt.show()
```

```python
plt.plot (x,y,linestyle=":", color = "#00FF00", lw = 08.0)
plt.show()
```

```python
plt.scatter(x,y, color = 'c')
plt.show()
```

```python
plt.bar(x,y, color = 'r')
plt.show()
```

```python
plt.scatter(x,y, color = 'c', marker = "p")
plt.show()
```
